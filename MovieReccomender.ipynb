# Finding Similar Movie -- ItemBasedCF

```python
# Import modules and read the file
import pandas as pd
import numpy as np
r_cols=['user_id','movie_id','rating']
ratings =pd.read_csv(r'data/ml-100k/u.data', names=r_cols,sep='\t',usecols=range(3),encoding='ISO-8859-1')
m_cols=['movie_id','title']
movies=pd.read_csv(r'data/ml-100k/u.item', names=m_cols,sep='|',usecols=range(2),encoding='ISO-8859-1')
```

```python
# View the data of ratings and users using head()
print(ratings.head())
movies.head()
```

```python
# Merge both into rating data
ratings=pd.merge(movies,ratings)
ratings.head()
```

```python
# Use pivot table to create matrix of users and movie they rated
user_ratings=ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
user_ratings.head()
```

```python
# Use pandas corr() method to compute the correlation score for every column pair in matrix
corr_matrix=user_ratings.corr()
corr_matrix.head()
```

```python
# to restrict our results to movies that lots of people rated together use min_periods argument to throw out our results
# where fewer than 100 users rated a given movie pair
# This will be done by using pearson method ( Pearson Correlation Coefficient )
corr_matrix=user_ratings.corr(method='pearson',min_periods=100)
corr_matrix.head()
```

```python
# View the rating at first index
my_ratings=user_ratings.loc[0].dropna()
my_ratings
```

```python
# 1. Go through each movie that user rated one at a atime and build a list of 
# all possible recommendations based on the movies similar to that ones user rated

similar_candidates=pd.Series()
for i in range(0,len(my_ratings.index)):
    print("Adding similarity for",my_ratings.index[i])
    sim=corr_matrix[my_ratings.index[i]].dropna()
    sim=sim.map(lambda x : x*my_ratings[i])
    similar_candidates=similar_candidates.append(sim)
    print(my_ratings.index[i])
    print(my_ratings[i])
    print('-------------------')

#sort the values

similar_candidates.sort_values(inplace=True ,ascending=False)
print(similar_candidates.head(10))
# 2. Retrieve the list of similar movies from our correlation matrix
    
    

# 3. Scale those correlation scores by how well user rated the movie they
# are similar to...

```

```python
# Use groupBy to add together the scores from movies that show up more than once, so they'll count more...
sim_cand=similar_candidates.groupby(similar_candidates.index).sum()
sim_cand.head()
sim_cand.sort_values(inplace=True ,ascending=False)
print(sim_cand.head(10))
```
